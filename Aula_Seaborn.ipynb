#Gr√°ficos Seaborn

```python
import pandas as pd
import seaborn as sns
dados = pd.read_csv('https://s3.amazonaws.com/caelum-online-public/1039-seaborn/material/tips.csv')
dados.head()
```

```python
dados.describe()
```

## Se as gorjetas maiores s√£o acompanhadas com valor da conta maior

```python
gorjeta = sns.scatterplot(x = 'total_bill', y = 'tip', data = dados)
```

```python
dados['porcentagem'] = (dados['tip'] / dados['total_bill']).round(4)
dados
```

```python
porcentagem_da_conta = sns.scatterplot(x = 'total_bill', y = 'porcentagem', data = dados)
```

```python
porcentagem_da_conta = sns.scatterplot(x = 'total_bill', y = 'porcentagem', data = dados)
```

```python
porcentagem_da_conta = sns.lmplot(x = 'total_bill', y = 'porcentagem', data = dados)
```
